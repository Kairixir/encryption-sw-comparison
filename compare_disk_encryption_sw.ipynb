```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tables = []
for i in range(1,6):
    tables.append(pd.read_csv(f"data/comparison_encryption_tab{i}.csv"))
```

```python
all = pd.DataFrame(columns=["Name"])
for table in tables:
    all = all.merge(table, on='Name', how='outer')
```

```python
all.info()
```

```python
df = all.copy(deep = True)
df = df[df["Maintained?"] != "No"]
#df = df[df["Mac OS X"] != "No"]
df = df[df["Licensing"] != "Proprietary"]
df = df[df["Linux"] != "No"]
df = df.reset_index(drop=True)
df.iloc[:, [0,1,2,3,4,8,9,10,11,21,23,24,25]]
```
